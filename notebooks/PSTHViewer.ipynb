```python
%matplotlib widget
%load_ext autoreload
%autoreload 2

from ephys.src.utils.viz import PSTHViewer, PSTHWidget
```

```python
subject = "GRB058"
session = "20260312_134952"

viewer = PSTHViewer(
    subject=subject,
    session=session,
    unit_criteria_id=1,
    pre_seconds=0.1,
    post_seconds=0.15,
    binwidth_ms=10,
    t_rise=0.001,
    t_decay=0.025,
    plot_type="raster",  # "raster" | "heatmap" | "psth"
)

widget = PSTHWidget(viewer)
widget.show()
```

```python
st_per_unit, align_ev, trial_df = viewer.compute()
```

```python
trial_df
```

```python
import numpy as np
stim = align_ev["stim"]
stim_ev = np.concatenate([[stim[0]], stim[1:][np.diff(stim) > 0.070]])
stim_ev.shape
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.eventplot(align_ev['stim'])
ax.set_xlim((0, 200));
```
