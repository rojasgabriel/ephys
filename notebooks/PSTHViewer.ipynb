```python
%matplotlib widget
%load_ext autoreload
%autoreload 2

from ephys.src.utils.viz import PSTHViewer, PSTHWidget
```

```python
subject = "GRB058"
session = "20260224_152424"

viewer = PSTHViewer(
    subject=subject,
    session=session,
    unit_criteria_id=1,
    pre_seconds=0.1,
    post_seconds=0.15,
    binwidth_ms=10,
    t_rise=0.001,
    t_decay=0.025,
    plot_type="raster",  # "raster" | "heatmap" | "psth"
)

widget = PSTHWidget(viewer)
widget.show()
```
